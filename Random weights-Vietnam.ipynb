```python
# import inbuilt modules
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# main module used for RDM calculations
import RDM_main
import rasterio as rio

import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

from random import random
import random
```

```python
def get_random_weights(seed=None):
    
    min_weight = .183333
    max_weight = .483333
    
    if seed:
        random.seed(seed)
    first_weight = random.uniform(min_weight,max_weight, )

    a = 1 - max_weight - first_weight
    b = 1 - min_weight - first_weight

    if a >= min_weight:
        min_weight2 = a
    else:
        min_weight2 = min_weight

    if b <= max_weight:
        max_weight2 = b
    else:
        max_weight2 = max_weight

    second_weight = random.uniform(min_weight2, max_weight2)
    third_weight =  1 - first_weight - second_weight

    weight_list = [first_weight, second_weight, third_weight]
    random.shuffle(weight_list)
    return weight_list
```
